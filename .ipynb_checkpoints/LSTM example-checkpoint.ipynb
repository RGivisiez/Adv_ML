```python
def fit_column_scaler_min_max(minval, maxval, dataframe):
    
    columns = dataframe.columns
    
    scaler = { column : MinMaxScaler(feature_range=(minval, maxval))
              for column in columns}
    
    scaler_fit = [scaler[column].fit_transform(dataframe[[column]])
                  for column in columns]
    
    for idx, column in enumerate(columns):
        dataframe[column] = scaler_fit[idx]
        
    return dataframe, scaler 
```

```python
df, scaler = fit_column_scaler_min_max(0, 1, df)
```

```python
def time_shift(df, lags, dropnan=True):
    
    '''
    Adiciona novas colunas deslocadas no tempo.
    Argumentos:
        df: Pandas DataFrame.
        lags: lista com o número de passos
               a ser deslocado.
    Retorna:
        Pandas DataFrame com as colunas deslocadas.
    '''
    lags.append(0)
    lags = -1 * np.sort(lags)
    
    columns = [str(df.name)] if type(df) is pd.Series else df.columns
    col = []
    
    for column in columns:
        for lag in lags: 
            
            if lag == 0:
                new_column = column 
                col.append(new_column)
            else:
                new_column = column + '_lag(' + str(-1 * lag) + ')'
                col.append(new_column)
                df[new_column] = df[column].shift(lag)
                
    if dropnan:
        df.dropna(inplace=True)
    
    return df[col]
```

# Divisão em janelas

```python
one_year = 365 * 24
```

```python
splits = TimeSeriesSplit(n_splits=4)

X = df.iloc[:one_year, :]

for train_idx, test_idx in splits.split(X):
    
    x_train = X.iloc[train_idx, 1:].values
    y_train = X.iloc[train_idx, 0].values
    
    x_test = X.iloc[test_idx, 1:].values
    y_test = X.iloc[test_idx, 0].values
        
    x_train = x_train.reshape((x_train.shape[0], 1, x_train.shape[1]))
    x_test = x_test.reshape((x_test.shape[0], 1, x_test.shape[1]))
    
    #Não sei se tem que criar o modelo toda hora.
    model = Sequential()
    model.add(LSTM(64, input_shape=(x_train.shape[1], x_train.shape[2])))
    model.add(Dropout(0.3))
    model.add(Dense(1))
    model.compile(loss='mae', optimizer='adam')

    
    history = model.fit(x_train, y_train, epochs=50, batch_size=72, validation_data=(x_test, y_test), verbose=0, shuffle=False)
    
    plt.figure(figsize=(7, 3))
    plt.plot(history.history['loss'], label='train')
    plt.plot(history.history['val_loss'], label='test')
    plt.legend()
    
    yhat = model.predict(x_test)

    yhat = scaler['pollution'].inverse_transform(yhat)

    y = scaler['pollution'].inverse_transform(y_test.reshape((-1,1)))
    
    rmse = np.sqrt(mean_squared_error(y, yhat))
    print('Test RMSE: %.3f' % rmse)
    
    #plt.figure(figsize=(15,3))
    #plt.plot(X.values, c='b', label='Completo')
    #plt.plot(train, c='r', label='Train')
    #plt.plot([ len(train) - 1 + i for i in range(test.shape[0])], test, 'g', label='Test')
    #plt.legend()
    
    plt.show()
```

```python
# Set de teste fora do treino. > que um ano.

test = df.iloc[one_year:, :]
x_test, y_test = test.iloc[:, 1:], test.iloc[:, 0]
x_test = x_test.values.reshape((x_test.shape[0], 1, x_test.shape[1]))
y_test = y_test.values
```

```python
yhat = model.predict(x_test)

yhat = scaler['pollution'].inverse_transform(yhat)

y = scaler['pollution'].inverse_transform(y_test.reshape((-1,1)))
    
rmse = np.sqrt(mean_squared_error(y, yhat))
print('Test RMSE: %.3f' % rmse)
```

```python
plt.figure(figsize=(15,10))
plt.plot(y, label='correct')
plt.plot(yhat, label='predicted')
plt.legend()
plt.xlim(0,2000)
plt.ylim(0,700)
plt.show()
```

# Walk forward (Rolling-Window)

```python
one_year = 365 * 24
```

```python
one_year, 6570+2190
```

```python
X = df[['pollution']].iloc[:one_year, :]

train_min = np.floor(one_year / 4).astype(np.int16)
train_max = one_year

for i in range(train_min, train_max):
    
    x_train = X.iloc[:i, :].values
    y_train = X.iloc[i:i+1, :].values
    
    batch_size = 1
    epoch = 1000
    
    model = Sequential()
    model.add(LSTM(64, batch_input_shape=(batch_size, x_train.shape[1], x_train.shape[2]), stateful=True))
    model.add(Dense(1))
    model.compile(loss='mae', optimizer='adam')    
    # recurrent_dropout=dropout
    for i in range(epoch):
        model.fit(x_train, y_train, epochs=1, batch_size=batch_size, verbose=0, shuffle=False)
        model.reset_states()
    
    
    #if (i % (train_min-1) == 0):
    #    plt.figure(figsize=(15,3))
    #    plt.plot(X.values, c='b', label='Completo')
    #    plt.plot(x_train, c='r', label='Train')
    #    plt.legend()
    
#plt.show()
```

```python
# make a one-step forecast
def forecast_lstm(model, batch_size, X):
    X = X.reshape(1, 1, len(X))
    yhat = model.predict(X, batch_size=batch_size)
    return yhat[0,0]
```
