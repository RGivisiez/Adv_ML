```python
import pandas as pd
import numpy as np
```

```python
left = pd.DataFrame({'key1': ['foo', 'foo', 'bar','extra1'],
 'key2': ['one', 'two', 'one','one'],
 'lval': [1, 2, 3,15]})
right = pd.DataFrame({'key1': ['foo', 'foo', 'bar', 'bar','extra'],
 'key2': ['one', 'one', 'one', 'two', 'one'],
 'rval': [4, 5, 6, 7, 10]})
print(left),print(' '), print(right)
```

```python
pd.merge(left, right, on=['key1'], how='inner')
```

```python
pd.merge(left, right, on=['key1'], how='outer')
```

```python
pd.merge(left, right, on=['key1'], how='left')
```

```python
pd.merge(left, right, on=['key1'], how='right')
```

Pensar no [key1, key2] como uma única coisa. Tipo, a combinação de key1,key2 é substituido por um número.

```python
pd.merge(left, right, on=['key1', 'key2'], how='outer')
```

```python
pd.merge(left, right, on=['key1', 'key2'], how='inner')
```

```python
df1 = pd.DataFrame(np.arange(6).reshape(3, 2), index=['a', 'b', 'c'], columns=['one', 'two'])
df2 = pd.DataFrame(5 + np.arange(4).reshape(2, 2), index=['a', 'c'], columns=['three', 'four'])
df3 = pd.DataFrame(1+np.arange(6).reshape(3, 2), index=['a', 'b', 'c'], columns=['one', 'two'])

print(df1), print(' '), print(df2), print(' '), print(df3)
```

```python
pd.concat([df1,df2], sort=False, axis=1, join='outer')
```

```python
pd.concat([df1,df3], sort=False, axis=0, join='outer')
```

```python
df1 = pd.DataFrame(np.random.randn(3, 4), columns=['a', 'b', 'c', 'd'])
df2 = pd.DataFrame(np.random.randn(2, 3), columns=['b', 'd', 'a'])
print(df1), print(' '), print(df2)
```
