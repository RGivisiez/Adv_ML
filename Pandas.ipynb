```python
import pandas as pd
import numpy as np
```

```python
left = pd.DataFrame({'key1': ['foo', 'foo', 'bar','extra1'],
 'key2': ['one', 'two', 'one','one'],
 'lval': [1, 2, 3,15]})
right = pd.DataFrame({'key1': ['foo', 'foo', 'bar', 'bar','extra'],
 'key2': ['one', 'one', 'one', 'two', 'one'],
 'rval': [4, 5, 6, 7, 10]})
print(left),print(' '), print(right)
```

```python
pd.merge(left, right, on=['key1'], how='inner')
```

```python
pd.merge(left, right, on=['key1'], how='outer')
```

```python
pd.merge(left, right, on=['key1'], how='left')
```

```python
pd.merge(left, right, on=['key1'], how='right')
```

Pensar no [key1, key2] como uma única coisa. Tipo, a combinação de key1,key2 é substituido por um número.

```python
pd.merge(left, right, on=['key1', 'key2'], how='outer')
```

```python
pd.merge(left, right, on=['key1', 'key2'], how='inner')
```

```python
a = [[0,1,3,1], [0,1,4,1], [0,1,4,2],[1,1,5,1],[1,1,5,1],[2,1,5,1],[2,1,4,4]]
train = pd.DataFrame(a, columns=['date_block_num', 'shop_id', 'item_id', 'item_cnt'])
test = pd.DataFrame([[1,1],[1,2]], columns=['shop_id', 'item_id'])
```

```python
train
```

```python
df_month = train.groupby(['date_block_num', 'shop_id', 'item_id']).sum()
df_month.rename(columns={'item_cnt':'sum_item_cnt_month'}, inplace=True)
df_month
```

```python
aux = train.groupby(['date_block_num', 'shop_id', 'item_id']).mean()
aux.rename(columns={'item_cnt':'mean_item_cnt_month'}, inplace=True)
df_month = pd.merge(df_month, aux, right_index=True, left_index=True)
df_month
```

```python
aux = train.groupby(['date_block_num', 'shop_id']).mean()
aux.drop(['item_id'], axis=1, inplace=True)
aux.rename(columns={'item_cnt':'mean_item_cnt_per_shop'}, inplace=True)
df_month = pd.merge(df_month.reset_index(), aux, on=['date_block_num', 'shop_id'], how='left')
df_month
```

```python
col = ['date_block_num', 'shop_id', 'item_id', 'sum_item_cnt_month', 'mean_item_cnt_per_shop']
df_month[col].pivot_table(index=['shop_id', 'item_id', 'mean_item_cnt_per_shop'], columns=['date_block_num'], fill_value=0)
```

```python
aux1 = df_month[col].pivot_table(index=['shop_id', 'item_id', 'mean_item_cnt_per_shop'], columns=['date_block_num'], fill_value=0)
aux2 = df_month[col].pivot_table(index=['shop_id', 'item_id', 'mean_item_cnt_per_shop'], columns=['date_block_num'], fill_value=0)['sum_item_cnt_month'].shift(1, axis=1)
aux2 = pd.concat([aux2], keys=['sum_item_cnt_month_lag(1)'], axis=1)
pd.concat([aux1, aux2], axis=1)
```
